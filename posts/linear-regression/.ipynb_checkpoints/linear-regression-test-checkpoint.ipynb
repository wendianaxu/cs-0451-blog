```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# data generation functions
def pad(X):
    return np.append(X, np.ones((X.shape[0], 1)), 1)

def LR_data(n_train = 100, n_val = 100, p_features = 1, noise = .1, w = None):
    if w is None: 
        w = np.random.rand(p_features + 1) + .2
    
    X_train = np.random.rand(n_train, p_features)
    y_train = pad(X_train)@w + noise*np.random.randn(n_train)

    X_val = np.random.rand(n_val, p_features)
    y_val = pad(X_val)@w + noise*np.random.randn(n_val)
    
    return X_train, y_train, X_val, y_val
```

```python
# generate data
n_train = 100
n_val = 100
p_features = 1
noise = 0.2

# create some data
X_train, y_train, X_val, y_val = LR_data(n_train, n_val, p_features, noise)

# plot it
fig, axarr = plt.subplots(1, 2, sharex = True, sharey = True)
axarr[0].scatter(X_train, y_train)
axarr[1].scatter(X_val, y_val)
labs = axarr[0].set(title = "Training", xlabel = "x", ylabel = "y")
labs = axarr[1].set(title = "Validation", xlabel = "x")
plt.tight_layout()
```

```python
def gradient(P, q, w):
        """
        Compute the gradient for the loss function of linear regression associated with w, X, and y. 
        """
        return P@w - q
```

```python
def fit(X, y, method = "analytic", alpha = 0.001, max_iter = 1000):
    X_ = pad(X)
    
    # analytic method
    if method == "analytic":
        w_hat = np.linalg.inv(X_.T@X_)@X_.T@y
        return w_hat
        
    # gradient descent method
    if method == "gradient":
        # determine n and p from X
        n = X_.shape[0] 
        p = X_.shape[1]
        
        # pick a random weight vector
        w = .5 - np.random.rand(p)
        
        # initialize prev_loss
        prev_loss = np.inf
       
        # calculate P and q for gradient descent
        P = X_.T@X_
        q = X_.T@y
        
        # main loop
        for i in np.arange(max_iter): 
            w -= 2*alpha*gradient(P, q, w)
            
            # compute loss
            new_loss = np.linalg.norm(X_@w - y, 2) ** 2
            
            # compute accuracy and append to history
            self.accuracy = self.score(X, y, self.w)
            self.score_history.append(self.accuracy)
            
            # check if loss hasn't changed and terminate if so
            if np.isclose(new_loss, prev_loss):          
                break
            else:
                prev_loss = new_loss
        
        return w
```

```python
def predict(X, w):
    return X@w

def score(X, y, w):
    y_hat = predict(X, w)
    y_bar = y.mean()
    c = 1 - ((y_hat - y) ** 2).sum() / ((y_bar - y) ** 2).sum()
    return c
```
